#key window moving factor

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        
        d = {}
        # creating hashmap
        for i in t:
            if i in d:
                d[i] += 1
            else:
                d[i] = 1
                
        count = len(d)
        i, j, n = 0, 0, len(s)
        ansl = float('inf')
        ans = ''
        
        while j < n:
            if s[j] in d:
                d[s[j]] -= 1
                if d[s[j]] == 0:
                    count -= 1   
            
            # increasing the window
            if count > 0:
                j += 1
                
            elif count == 0:      
                # removing extra chars from start i.e. making window size smaller from left
                while i <= j  and  (s[i] not in d  or  d[s[i]] < 0):
                    if s[i] in d:
                        d[s[i]] += 1
                    i+=1
                    
                # comparing with current answer and updating
                if (j-i+1) < ansl:
                    ansl = j-i+1
                    ans = s[i:j+1]
                    
                # sliding the window
                if i <= j and s[i] in d:
                    d[s[i]] += 1
                    if d[s[i]] == 1:
                        count += 1
                i += 1
                j += 1
                
        return ans
```
